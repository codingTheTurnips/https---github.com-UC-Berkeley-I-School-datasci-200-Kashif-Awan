````python
# QP1-1 Grading Tag:

# TODO: ALL OF YOUR CODE TO GRADE NEEDS TO EXIST IN THIS CELL

import random

class LocalView:
    def __init__(
        self,
        can_move_north: bool = False,
        can_move_south: bool = False,
        can_move_east: bool = False,
        can_move_west: bool = False,
    ):
        self.can_move_north = can_move_north
        self.can_move_south = can_move_south
        self.can_move_east = can_move_east
        self.can_move_west = can_move_west

    @property
    def available_directions(self):
        directions = []
        if self.can_move_north: directions.append("north")
        if self.can_move_east:  directions.append("east")
        if self.can_move_south: directions.append("south")
        if self.can_move_west:  directions.append("west")
        return directions

    @available_directions.setter
    def available_directions(self, directions: list):
        self.can_move_north = "north" in directions
        self.can_move_east  = "east"  in directions
        self.can_move_south = "south" in directions
        self.can_move_west  = "west"  in directions

    @property
    def move_directions(self):
        return ["stay"] + self.available_directions

    def allow(self, direction: str):
        setattr(self, f"can_move_{direction.lower()}", True)

    def block(self, direction: str):
        setattr(self, f"can_move_{direction.lower()}", False)

    def __repr__(self):
        dirs = self.available_directions
        return f"LocalView(can move: {', '.join(dirs) if dirs else 'none'})"

    def __str__(self):
        return str(self.available_directions)
    
class Animal:
    """main class for Animal, has properties that all subclasses use"""
    def __init__(
            self
    ):
        self.icon = None
    def icon(self,value):
        if len(value) != 1:
            raise Exception("icon must be a single character.")
        self.icon = value
    def move(self, lv: LocalView):
        """What direction can the animal move"""
        if lv == None:
            self.move = LocalView()
        else:
            self.move = lv
        raise NotImplementedError(f"{self.__class__.__name__} must implement move()")


    def __str__(self):
        return self.icon if self.icon else ""



class hamster(Animal):
    def __init__(self):
        super().__init__()
        self.icon = '\U0001F439' 
    def move(self, local_view = LocalView(can_move_east=True, can_move_north=True, can_move_south=True, can_move_west=True)) -> str:
        #remove  = LocalView(can_move_east=True, can_move_north=True, can_move_south=True, can_move_west=True <- check the logic
        options = local_view.available_directions #can be removed
        #object is getting stored, and being reused. -- enforce the movment in the body
        return random.choice(options) if options else "stay"
    
class butterfly(Animal):
    def __init__(self):
        super().__init__()
        self.icon = '\U0001F439' 
    def move(self, local_view = LocalView(can_move_east=True, can_move_north=True, can_move_south=True, can_move_west=True)) -> str:
        options = local_view.available_directions
        return random.choice(options) if options else "stay"

class Zoo:
    """Rectangular grid enclosure that tracks animal positions."""

    def __init__(self, height: int, width: int):
        self.height = height
        self.width = width
        self._animal_to_pos = {}
        self._pos_to_animal = {}

    def add_animal(self, animal, row: int, col: int):
        if not (0 <= row < self.height and 0 <= col < self.width):
            raise ValueError(f"Position ({row},{col}) is outside the zoo grid.")
        if (row, col) in self._pos_to_animal:
            raise ValueError(f"Position ({row},{col}) is already occupied.")
        self._animal_to_pos[animal] = (row, col)
        self._pos_to_animal[(row, col)] = animal

    def position_of_animal(self, animal):
        return self._animal_to_pos.get(animal, None)

    def animal_at_position(self, row: int, col: int):
        return self._pos_to_animal.get((row, col), None)

    def _build_local_view(self, row: int, col: int) -> LocalView:
        """Build a LocalView for a position based on grid bounds and occupancy."""
        return LocalView(
            can_move_north=row > 0 and (row - 1, col) not in self._pos_to_animal,
            can_move_south=row < self.height - 1 and (row + 1, col) not in self._pos_to_animal,
            can_move_east= col < self.width - 1 and (row, col + 1) not in self._pos_to_animal,
            can_move_west= col > 0      and (row, col - 1) not in self._pos_to_animal,
        )

    def _apply_move(self, animal, direction: str):
        """hope the animal moves correctly"""
        row, col = self._animal_to_pos[animal]
        cardinalMove = {
            "north": (-1,  0), #move up
            "south": ( 1,  0), #move down
            "east":  ( 0,  1), # move right
            "west":  ( 0, -1), #move left
            "stay":  ( 0,  0),
        }
        rowMove, columnMove = cardinalMove[direction]
        #why can't i just add two tuples, it keeps breaking
        #
        new_pos = (row + rowMove, col + columnMove)

        if new_pos != (row, col):
            del self._pos_to_animal[(row, col)]
            self._animal_to_pos[animal] = new_pos
            self._pos_to_animal[new_pos] = animal

    def time_step(self):
        """Give every animal one chance to move, one at a time."""
        for animal in list(self._animal_to_pos):
            row, col = self._animal_to_pos[animal]
            local_view = self._build_local_view(row, col)
            direction = animal.move(local_view)
            self._apply_move(animal, direction)
            print(f"Animal at ({row},{col})")
            print(f"  can_move_north: {local_view.can_move_north}")
            print(f"  can_move_south: {local_view.can_move_south}")
            print(f"  can_move_east:  {local_view.can_move_east}")
            print(f"  can_move_west:  {local_view.can_move_west}")

    def __repr__(self):
        return f"Zoo(height={self.height}, width={self.width}, animals={len(self._animal_to_pos)})"
````

````python
import random

def test_animal_moves_around_zoo():
    results = []

    # --- Test 1: Animal starts at correct position ---
    zoo = Zoo(10, 10)
    dog = hamster()
    zoo.add_animal(dog, 5, 5)
    assert zoo.position_of_animal(dog) == (5, 5), "Animal should start at (5,5)"
    results.append("✅ Test 1 passed: animal placed at correct position")

    # --- Test 2: Animal moves after time_step ---
    random.seed(42)
    zoo = Zoo(10, 10)
    dog = hamster()
    zoo.add_animal(dog, 5, 5)
    zoo.time_step()
    new_pos = zoo.position_of_animal(dog)
    assert new_pos != (5, 5), "Dog should have moved from (5,5)"
    results.append(f"✅ Test 2 passed: dog moved to {new_pos}")

    # --- Test 3: Animal stays within grid bounds after many steps ---
    zoo = Zoo(10, 10)
    dog = hamster()
    zoo.add_animal(dog, 5, 5)
    for _ in range(100):
        zoo.time_step()
        row, col = zoo.position_of_animal(dog)
        assert 0 <= row < zoo.height, f"Row {row} out of bounds"
        assert 0 <= col < zoo.width,  f"Col {col} out of bounds"
    results.append("✅ Test 3 passed: animal stayed in bounds after 100 steps")

    # --- Test 4: Animal forced to stay when all directions blocked ---
    zoo = Zoo(3, 3)
    animals = [hamster(), butterfly(), hamster(), hamster(), hamster(), hamster(), hamster(), butterfly(), butterfly()]
    positions = [(r, c) for r in range(3) for c in range(3)]
    for animal, pos in zip(animals, positions):
        zoo.add_animal(animal, *pos)
    center_animal = zoo.animal_at_position(1, 1)
    zoo.time_step()
    assert zoo.position_of_animal(center_animal) == (1, 1), "Fully surrounded animal should stay"
    results.append("  Test 4 passed: surrounded animal stayed in place")

    # --- Test 5: Two animals never occupy the same cell ---
    zoo = Zoo(5, 5)
    dog1, dog2 = hamster(), hamster()
    zoo.add_animal(dog1, 0, 0)
    zoo.add_animal(dog2, 0, 1)
    for _ in range(50):
        zoo.time_step()
        pos1 = zoo.position_of_animal(dog1)
        pos2 = zoo.position_of_animal(dog2)
        assert pos1 != pos2, f"Two animals occupy same cell {pos1}!"
    results.append("  Test 5 passed: no two animals shared a cell over 50 steps")

    # --- Test 6: Animal in corner can only move along edges ---
    zoo = Zoo(5, 5)
    dog = hamster()
    zoo.add_animal(dog, 0, 0)
    for _ in range(20):
        zoo.time_step()
        row, col = zoo.position_of_animal(dog)
        assert 0 <= row < zoo.height
        assert 0 <= col < zoo.width
    results.append("  Test 6 passed: corner animal stayed in bounds")

    # --- Test 7: add_animal raises on out-of-bounds position ---
    zoo = Zoo(5, 5)
    dog = hamster()
    try:
        zoo.add_animal(dog, 10, 10)
        results.append("❌ Test 7 failed: should have raised ValueError")
    except ValueError:
        results.append("  Test 7 passed: out-of-bounds position raised ValueError")

    # --- Test 8: add_animal raises on occupied position ---
    zoo = Zoo(5, 5)
    dog1, dog2 = hamster(), hamster()
    zoo.add_animal(dog1, 2, 2)
    try:
        zoo.add_animal(dog2, 2, 2)
        results.append("❌ Test 8 failed: should have raised ValueError")
    except ValueError:
        results.append("  Test 8 passed: occupied position raised ValueError")

    # --- Summary ---
    print("\n".join(results))
    print(f"\n{len(results)}/8 tests passed.")


test_animal_moves_around_zoo()
````
